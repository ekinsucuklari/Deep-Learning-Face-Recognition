```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model

# Modeli yükle
model = load_model('onlydropout.h5')  # Model dosya adınızı buraya ekleyin

# Etiketler
labels = {0: 'happy', 1: 'sad'}  # Sınıf etiketlerinize göre ayarlayın

# Kamera başlat
cap = cv2.VideoCapture(0)

# Başlangıç zamanını al
start_time = cv2.getTickCount()

while True:
    # Kameradan görüntü yakala
    ret, frame = cap.read()
    if not ret:
        break
    
    # Görüntüyü grayscale yap ve boyutlandır
    
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    resized_frame = cv2.resize(gray_frame, (48, 48))
    
    # Görüntüyü modelin girişine uygun hale getir
    #batch boyutu 1 foto old için, görüntü yğksekliği, genişliği, kanal sayısı BGR old için 1

    reshaped_frame = np.reshape(resized_frame, (1, 48, 48, 1))
    reshaped_frame = reshaped_frame / 255.0  # Normalizasyon
    
    # Tahmin yap
    prediction = model.predict(reshaped_frame)
    predicted_class = int(np.round(prediction)[0][0])

    # Tahmini ekrana yazdır
    cv2.putText(frame, labels[predicted_class], (20, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    
    # Görüntüyü göster
    cv2.imshow('Emotion Detection', frame)

    # Çıkış için 'q' tuşuna basın veya 60 saniye sonra dur
    end_time = cv2.getTickCount()
    elapsed_time = (end_time - start_time) / cv2.getTickFrequency()
    if elapsed_time >= 60 or cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Kamera kapat
cap.release()
cv2.destroyAllWindows()
```
